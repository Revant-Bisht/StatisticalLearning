```python
import numpy as np
# Q4 (graded)
A_1 = np.array([
    [1,2,3],
    [3,2,1],
    [2,2,1]
], dtype=np.float64)

detA_1 = np.linalg.det(A_1)

print(f"Determinant of matrix A_1: {detA_1:.2f}")
```

```python
# Q5 (graded)
A_2 = np.array([
    [1,2,3],
    [0,2,2],
    [1,4,5]
], dtype=np.float64)

detA_1 = np.linalg.det(A_2)

print(f"Determinant of matrix A_2: {detA_1:.2f}")
```
